```python
import numpy as np
from trustthebacktest.overfitting.pbo import estimate_pbo

T = 252 * 5
N = 50
rng = np.random.default_rng(0)

base = rng.normal(0.0002, 0.01, size=(T, 1))
drifts = rng.normal(0.0, 0.0001, size=(1, N))
noise = rng.normal(0.0, 0.01, size=(T, N))

R = base + drifts + noise

res = estimate_pbo(R, S=16, K=8)
res.pbo, res.n_folds, res.n_strategies
```

```python
from trustthebacktest.plotting.plots import plot_lambda_hist, plot_rank_hist

plot_lambda_hist(res)
plot_rank_hist(res)
```
